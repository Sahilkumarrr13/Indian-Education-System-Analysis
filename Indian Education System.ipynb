# Importing Libraries:

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Importing Dataset:

```python
dropout = pd.read_csv('./Dataset/dropout-ratio-2012-2015.csv')
dropout.head()
```

```python
# All the columns in our dropout dataframe

dropout.columns
```

```python
# Looking for Missing Values

dropout.isna().sum()
```

We can see there are no null values.

# Exploratory Data Analysis:

We can start by looking at the data statewise.

Let's look for West Bengal

```python
dropout['State_UT'].value_counts()
```

```python
data_wb = dropout[dropout['State_UT'] == 'West Bengal']
data_wb.head()
```

```python
# Statistical Information about the dataframe:

data_wb.info()
```

But we want to convert these columns into float

```python
for i in data_wb.columns[2:]:
    dropout[i] = dropout[i].astype('float')

dropout.info()
```

We can see that we dont have missing values but we do have inconsistent values like NR.

Let us look at these values.

```python
for i in data_wb.columns[2:]:
    print(i, '\n', dropout[i].value_counts(), end = " ")
```

We can see we have NR as Inconsistent values.

```python
dropout.replace('NR', '0', inplace=True)
```

```python
dropout['Upper Primary_Boys'].apply(lambda x: x == 'Uppe_r_Primary').sum()
```

```python
dropout['Upper Primary_Boys'].replace('Uppe_r_Primary', '0', inplace=True)
```

```python
# We can now convert it to float.

for i in data_wb.columns[2:]:
    dropout[i] = dropout[i].astype('float')

dropout.info()
```

# Data Visualisation:

```python
dropout.head()
```

```python
# Ploting bar plot for Drop Out of Primary Girls

plt.figure(figsize=(12, 9))
plt.bar( dropout['year'], dropout['Primary_Girls'])
plt.title("Yearly Drop Out Population of Primary Girls")
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

```python
# Ploting bar plot for Drop Out of Primary Girls

plt.figure(figsize=(12, 9))
plt.bar( dropout['year'], dropout['Primary_Boys'])
plt.title("Yearly Drop Out Population of Primary Boys")
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

```python
# Outliers Detection using BoxPlot

plt.figure(figsize=(20, 15))

plt.subplot(4, 4, 1)
sns.boxplot(x = 'HrSecondary_Boys', data = dropout)

plt.subplot(4,4,2)
sns.boxplot(x = 'HrSecondary_Girls', data = dropout)

plt.subplot(4,4,3)
sns.boxplot(x = 'Upper Primary_Boys', data = dropout)

plt.subplot(4,4,4)
sns.boxplot(x = 'Upper Primary_Girls', data = dropout)

plt.subplot(4,4,5)
sns.boxplot(x = 'Secondary _Boys', data = dropout)

plt.subplot(4,4,6)
sns.boxplot(x = 'Secondary _Girls', data = dropout)
plt.show()
```

```python
# Removing Outliers using clip function

dropout['HrSecondary_Boys'].clip(lower=dropout['HrSecondary_Boys'].quantile(0.05), 
                                 upper=dropout['HrSecondary_Boys'].quantile(0.95), inplace=True)

dropout['HrSecondary_Girls'].clip(lower=dropout['HrSecondary_Girls'].quantile(0.05), 
                                 upper=dropout['HrSecondary_Girls'].quantile(0.95), inplace=True)

dropout['Secondary _Boys'].clip(lower=dropout['Secondary _Boys'].quantile(0.05), 
                                 upper=dropout['Secondary _Boys'].quantile(0.95), inplace=True)

dropout['Secondary _Girls'].clip(lower=dropout['Secondary _Girls'].quantile(0.05), 
                                 upper=dropout['Secondary _Girls'].quantile(0.95), inplace=True)

dropout['Upper Primary_Boys'].clip(lower=dropout['Upper Primary_Boys'].quantile(0.05), 
                                 upper=dropout['Upper Primary_Boys'].quantile(0.95), inplace=True)

dropout['Upper Primary_Girls'].clip(lower=dropout['Upper Primary_Girls'].quantile(0.05), 
                                 upper=dropout['Upper Primary_Girls'].quantile(0.95), inplace=True)
```

```python
# Outliers Detection using BoxPlot

plt.figure(figsize=(20, 15))

plt.subplot(4, 4, 1)
sns.boxplot(x = 'HrSecondary_Boys', data = dropout)

plt.subplot(4,4,2)
sns.boxplot(x = 'HrSecondary_Girls', data = dropout)

plt.subplot(4,4,3)
sns.boxplot(x = 'Upper Primary_Boys', data = dropout)

plt.subplot(4,4,4)
sns.boxplot(x = 'Upper Primary_Girls', data = dropout)

plt.subplot(4,4,5)
sns.boxplot(x = 'Secondary _Boys', data = dropout)

plt.subplot(4,4,6)
sns.boxplot(x = 'Secondary _Girls', data = dropout)
plt.show()
```
